```python
import pandas as pd
```

```python

# Path to your input .parquet file
input_parquet_path = '../data/Date_final_dataset_balanced_float32.parquet'

# Path to your output .csv file
output_csv_path = 'output_file.csv'
```

```python

# Read the Parquet file
df = pd.read_parquet(input_parquet_path)
df
```
